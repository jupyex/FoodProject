```python
# Import uuid
import uuid

# Import Operating System
import os
```

```python
labels = ['rice-white-steamed', 'char-siu', 'beef-round-steak', 'chicken-wing', 'apple','banana','scrambled-egg','broccoli','carrot','mushroom-white']
number_imgs = 5
```

```python
IMAGES_PATH = os.path.join('Tensorflow', 'workspace', 'images', 'collectedimages')
```

```python
if not os.path.exists(IMAGES_PATH):
    if os.name == 'posix':
        !mkdir -p {IMAGES_PATH}
    if os.name == 'nt':
         !mkdir {IMAGES_PATH}
for label in labels:
    path = os.path.join(IMAGES_PATH, label)
    if not os.path.exists(path):
        !mkdir {path}
```

```python
# Iterate over the files in the directory
for label in labels:
    directory = os.path.join(IMAGES_PATH, label)
    for filename in os.listdir(directory):
        new_filename = '{}.jpg'.format(str(uuid.uuid1()))
        old_filepath = os.path.join(directory, filename)
        new_filepath = os.path.join(directory, new_filename)
        os.rename(old_filepath, new_filepath)
        print('Renamed {} to {}'.format(filename, new_filename))
```
